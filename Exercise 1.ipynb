```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt

NUMBER = 10001
MIN = 0.01
MAX = 500

x = np.linspace(MIN, MAX, NUMBER)
print(x)
fx = 2 / x**2
print(fx)
fx[x < 1] = 0
fx[x > 2] = 0
#fx = np.divide(2, x**2, where=(x > 1) & (x <= 2), out=np.zeros(NUMBER))
#fx = np.where(x > 2, 0, np.where(x < 1, 0, 2 / x**2))

fig = plt.gcf()
fig.set_size_inches(18, 5)
plt.scatter(x, fx)
plt.show()
```

```python
np.sum(fx) / (NUMBER + 1) * (MAX - MIN)
```

```python
#Fx = np.divide(2 * (x - 1), x, where=(x > 1) & (x <= 2), out=np.where(x < 2, np.zeros(10001), np.ones(10001)))
Fx = 2 * (x - 1) / x
Fx[x < 1] = 0
Fx[x > 2] = 1
Fxnum = np.cumsum(fx / NUMBER * (MAX - MIN))
fig = plt.gcf()
fig.set_size_inches(18, 5)
plt.plot(x, Fx)
plt.plot(x, Fxnum)
plt.show()
Cx = -2 / (x - 2)
Cx[x < 0] = 0
Cx[x > 1] = np.nan

r = np.random.uniform(0, 1, size=100000)
t = -2 / (r - 2)
plt.hist(t, bins=300)

plt.show()
```

```python
mu = 0
sigma = 1
lnormx = 1 / (x * np.sqrt(2 * np.pi)) * np.exp(-((np.log(x) - mu)**2 / (2 * sigma**2)))
```

```python
plt.plot(x, lnormx, c='green')
fig = plt.gcf()
fig.set_size_inches(18, 5)
plt.vlines(np.exp(1/2), 0, np.max(lnormx), color='orange')
plt.vlines(np.exp(-1), 0, np.max(lnormx), color='blue')
plt.show()
```

```python
np.sum(x * lnormx * (MAX - MIN) / NUMBER)**2
```
