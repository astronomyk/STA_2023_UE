```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from astropy.io import ascii

data = ascii.read('data.dat')
print(data[0])

plt.bar(data['mJy'], data['G'])
Gcum = np.cumsum(data['G']) / 386
Rcum = np.cumsum(data['R']) / 290
plt.show()


plt.bar(data['mJy'], np.cumsum(data['G']) / 386)
plt.bar(data['mJy'], np.cumsum(data['R']) / 290, alpha=0.5)
print(np.argmax(Rcum - Gcum) - 15, np.max(Rcum - Gcum))
plt.show()
```
